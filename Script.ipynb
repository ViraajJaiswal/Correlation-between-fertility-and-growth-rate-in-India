```python
import pandas as pd
from matplotlib import pyplot as plt

fertility_rates = pd.read_csv('fertility_rates.csv', encoding='latin1', dtype=str)
ind_fertility_rate = fertility_rates[fertility_rates['Country Code'] == 'IND']

ind_fertility_rate_plot = ind_fertility_rate.iloc[:, 3:64].T
ind_fertility_rate_plot = ind_fertility_rate_plot.apply(pd.to_numeric)

growth_rates = pd.read_csv('growth_rates.csv', encoding='utf-8', dtype=str)
ind_growth_rate = growth_rates[growth_rates['Country Code'] == 'IND']

ind_growth_rate_plot = ind_growth_rate.iloc[:, 5:66].T
ind_growth_rate_plot = ind_growth_rate_plot.apply(pd.to_numeric)

years = fertility_rates.columns[3:64]
years_list = [str(year) for year in years]

plt.plot(years_list, ind_fertility_rate_plot, color = 'green')
plt.plot(years_list, ind_growth_rate_plot, color = 'red')
plt.xticks(years_list[::10])

plt.xlabel('Years →')
plt.ylabel('Fertility and Growth Rate (%) →')
plt.title('Fertility and Growth Rates in India (1961-2021)')
plt.legend(["Fertility Rate", "Growth Rate"])
plt.tight_layout()

plt.savefig('Graph.jpg')
plt.show() 
```
